```python
#init
tok = "6e600273078e0ef22636298e3bed37f38c577d8829ba250731368234"
pro = ts.pro_api(tok)
```

```python
df = pro.index_daily(ts_code='000001.SH', start_date='20180101', end_date='20191231')
```

```python
df1 = pro.index_daily(ts_code='000052.SH', start_date='20180101', end_date='20191231')
```

```python
p = df['close']

c= Line()

c.add_xaxis(df['trade_date']
           )
c.add_yaxis("上证指数", df['close'].values,
            label_opts = opts.LabelOpts(is_show=False), 
            is_smooth = False,
           )
c.add_yaxis("上证50指数", df1['close'].values,
            label_opts = opts.LabelOpts(is_show=False), 
            is_smooth = False,
           )


#c.set_series_opts(label_opts=opts.LabelOpts(is_show=True))
c.set_global_opts(yaxis_opts=opts.AxisOpts(name = "y",
                                           is_show = True, 
                                           position = "top",
                                           #split_number = 10,
                                           min_ = 2000,
                                           ),
                 xaxis_opts = opts.AxisOpts(name = "20180101 - 20191231",
                                            type_ = "category",
                                            is_show = True,
                                            name_location = "center",
                                            name_gap = 30,
                                            axispointer_opts = opts.AxisPointerOpts(is_show = True, type_="line"),
                                            min_interval = 1
                                            
                 )
                 )

c.render_notebook()
```
